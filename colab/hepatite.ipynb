```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Aulas/DSSaude/code/hepatitis-classifier/colab/HepatitisCdata.csv")
X, y = df.iloc[:,2:], df.iloc[:,1:2]
```

```python
X
```

```python
dummies = pd.get_dummies(X[['Sex']])
X = pd.concat([X, dummies], axis=1)

X = X.drop(X[['Sex']], axis=1)
```

```python
X['ALB'] = X['ALB'].fillna(X['ALB'].mean())
X['ALP'] = X['ALP'].fillna(X['ALP'].mean())
X['ALT'] = X['ALT'].fillna(X['ALT'].mean())
X['CHOL'] = X['CHOL'].fillna(X['CHOL'].mean())
X['PROT'] = X['PROT'].fillna(X['PROT'].mean())
```

```python
X
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

rna = MLPClassifier(random_state=103)
rna = rna.fit(X, y)
y_predict = rna.predict(X)
print(accuracy_score(y,y_predict))
print(classification_report(y,y_predict))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score, make_scorer
from sklearn.model_selection import LeaveOneOut

scoring = {
      'precision_0': make_scorer(custom_precision_0),
      'recall_0':  make_scorer(custom_recall_0),
      'precision_1': make_scorer(custom_precision_1),
      'recall_1':  make_scorer(custom_recall_1),
      'accuracy': 'accuracy'
      }
model = MLPClassifier(random_state=103)
scores = cross_validate(model, X, y, cv=10, scoring=scoring)
```

```python
head_fmt = "{:>{width}s} " + " {:>9}" * 2
print(head_fmt.format("", *("precision","recall"), width=12))

row_fmt = "{:>{width}s} " + " {:>9.2f}" * 2 + ""

print(row_fmt.format(*("0",round(scores['test_precision_0'].mean(),2), round(scores['test_recall_0'].mean(),2)), width=10, digits=2))
print(row_fmt.format(*("1",round(scores['test_precision_1'].mean(),2), round(scores['test_recall_1'].mean(),2)), width=10, digits=2))
print(f"accuracy: {round(scores['test_accuracy'].mean(),2)}")
```

Oversampling

```python
from sklearn.model_selection import KFold

from imblearn.over_sampling import SMOTE

kf = KFold(n_splits=10, random_state = 103, shuffle=True)

p1 = []
p0 = []
r1 = []
r0 = []
accuracy = []

for fold, (train_index, test_index) in enumerate(kf.split(X), 1):
    X_train = X.iloc[train_index]
    y_train = y.iloc[train_index]  # Based on your code, you might need a ravel call here, but I would look into how you're generating your y
    X_test = X.iloc[test_index]
    y_test = y.iloc[test_index]  # See comment on ravel and  y_train
    sm = SMOTE(random_state = 103)
    X_train_oversampled, y_train_oversampled = sm.fit_resample(X_train, y_train)
    model = MLPClassifier(random_state=103)
    model.fit(X_train_oversampled, y_train_oversampled )  
    y_pred = model.predict(X_test)
    p1.append(precision_score(y_test,y_pred, pos_label=1))
    r1.append(recall_score(y_test,y_pred, pos_label=1))
    p0.append(precision_score(y_test,y_pred, pos_label=0))
    r0.append(recall_score(y_test,y_pred, pos_label=0))
    accuracy.append(accuracy_score(y_test,y_pred))
```

```python
from statistics import mean
head_fmt = "{:>{width}s} " + " {:>9}" * 2
print(head_fmt.format("", *("precision","recall"), width=12))


row_fmt = "{:>{width}s} " + " {:>9.2f}" * 2 + ""

print(row_fmt.format(*("0",round(mean(p0),2), round(mean(r0),2)), width=10, digits=2))
print(row_fmt.format(*("1",round(mean(p1),2), round(mean(r1),2)), width=10, digits=2))
print(f"accuracy: {round(mean(accuracy),2)}")
```
